```python
import numpy as np
import pandas as pd
import joblib
```

```python
iso_model = joblib.load("isolation_forest_model.pkl")
rf_model = joblib.load("random_forest_model.pkl")

print("Models Loaded")
```

```python
FEATURES = [
    "mean_cpu",
    "std_cpu",
    "min_cpu",
    "max_cpu",
    "delta_cpu",
    "cpu_trend",
    "cpu_volatility",
    "mean_memory",
    "std_memory",
    "memory_trend",
    "mean_requests",
    "request_spike_count",
    "throughput_delta",
    
    "cost_delta",
    "unit_economics_ratio"
]
```

```python

test_df = pd.read_csv(r"C:\Users\Athul VR\OneDrive\Desktop\ML Model\featuring\randomdata_final.csv")

X_test = test_df[FEATURES].copy()
```

```python
import pandas as pd

def run_ml_pipeline(feature_vector):

    X = pd.DataFrame(
        [feature_vector],
        columns=FEATURES
    )

    raw_score = iso_model.decision_function(X)

    risk_score = 1 / (1 + np.exp(3 * raw_score))

    is_anomaly = risk_score[0] > THRESHOLD

    if is_anomaly:
        incident = rf_model.predict(X)[0]
        confidence = np.max(rf_model.predict_proba(X))
    else:
        incident = "Normal"
        confidence = 0.95

    return risk_score[0], is_anomaly, incident, confidence
```

```python
sample = X_test.iloc[0].values   # use your test feature row

risk, anomaly, incident, conf = run_ml_pipeline(sample)

print("Risk:", risk)
print("Anomaly:", anomaly)
print("Incident:", incident)
print("Confidence:", conf)
```
