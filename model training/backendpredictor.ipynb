```python
def predict_full_stack(iso_model, rf_model, feature_vector):

    # Convert to dataframe
    X = feature_vector.reshape(1, -1)

    # ---------- Isolation ----------
    risk_score = get_anomaly_score(iso_model, X)[0]
    is_anomaly = risk_score > 0.6

    # ---------- Random Forest ----------
    if is_anomaly:
        incident, confidence = get_rf_prediction_with_confidence(rf_model, X)
    else:
        incident = "Normal"
        confidence = 1 - risk_score

    # ---------- Indicators ----------
    feature_dict = dict(zip(COMMON_FEATURES, feature_vector))
    indicators = generate_indicators(feature_dict)

    return {
        "anomaly_score": float(risk_score),
        "is_anomaly": bool(is_anomaly),
        "incident_type": incident,
        "confidence": float(confidence),
        "indicators": indicators,
        "features": feature_dict
    }
```
